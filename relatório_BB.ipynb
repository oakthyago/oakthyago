```python
import seaborn as sns
import matplotlib.pyplot as plt


employee_dict = {
    "F1367519": "ARTHUR LIRA DA SILVA",
    "F1669799": "BONIEK CICERO DA SILVA",
    "F1833287": "CARLOS HENRIQUE VALENCA ARAUJO DA SILVA",
    "F2125367": "CLEYTON DA SILVA GOMES",
    "F2253189": "DANIEL BIDO CARVALHO",
    "F3296093": "FLAVIA APOLINARIO OLIVEIRA FARIAS",
    "F3295818": "FLAVIANO DA SILVA",
    "F3417358": "FRANCISCO GENILSON MARINHO MEDEIROS",
    "F5721682": "JOSE NAZARENO DE OLIVEIRA",
    "F6519319": "LUIZ GUEDES MONTEIRO NETO",
    "F8030919": "PAULA PACHECO CARDOSO",
    "F8620357": "RITA DE CASSIA CABRAL DE MELO",

    "F9342321": "THYAGO BORGES DE CARVALHO",
    "F9476713": "VERONICA MARIA PACHECO DE ARAUJO MOURA",
    "F9916863": "WLADIMIR FERNANDES DA SILVA"
}


# Initialize efficacy list
eficacia = []

# Define distinct colors for each employee
distinct_colors = {
    "ARTHUR LIRA DA SILVA": "#FF0000",   # Red
    "BONIEK CICERO DA SILVA": "#00FF00", # Green
    "CARLOS HENRIQUE VALENCA ARAUJO DA SILVA": "#0000FF", # Blue
    "CLEYTON DA SILVA GOMES": "#FFFF00", # Yellow
    "DANIEL BIDO CARVALHO": "#FF00FF",  # Magenta
    "FLAVIA APOLINARIO OLIVEIRA FARIAS": "#00FFFF", # Cyan
    "FLAVIANO DA SILVA": "#800080", # Purple
    "FRANCISCO GENILSON MARINHO MEDEIROS": "#FFA500", # Orange
    "JOSE NAZARENO DE OLIVEIRA": "#A52A2A", # Brown
    "LUIZ GUEDES MONTEIRO NETO": "#808000", # Olive
    "PAULA PACHECO CARDOSO": "#008080", # Teal
    "RITA DE CASSIA CABRAL DE MELO": "#800000", # Maroon

    "THYAGO BORGES DE CARVALHO": "#DC143C", # Crimson
    "VERONICA MARIA PACHECO DE ARAUJO MOURA": "#2E8B57", # Sea Green
    "WLADIMIR FERNANDES DA SILVA" : "#4682B4"
}

# Calculate metrics for each employee
eficacia = []

for emp_id, emp_name in employee_dict.items():
    if emp_id in df["Funcionário"].values:
        filtered_df = df[df["Funcionário"] == emp_id]

        # Calculate percentage and absolute values
        x = filtered_df["Atendimento"].sum() / len(filtered_df) * 100
        x1 = filtered_df["Atendimento"].sum()
        y = filtered_df["Cordialidade"].sum() / len(filtered_df) * 100
        y1 = filtered_df["Cordialidade"].sum()
        z = filtered_df["Dominio"].sum() / len(filtered_df) * 100
        z1 = filtered_df["Dominio"].sum()

        eficacia.append([emp_name, x, y, z, x1, y1, z1])

# Convert data to DataFrames
eficacia_df_percent = pd.DataFrame(
    eficacia, columns=["Funcionário", "Atendimento", "Cordialidade", "Dominio", "Atendimento_Abs", "Cordialidade_Abs", "Dominio_Abs"]
)[["Funcionário", "Atendimento", "Cordialidade", "Dominio"]]

eficacia_df_abs = pd.DataFrame(
    eficacia, columns=["Funcionário", "Atendimento", "Cordialidade", "Dominio", "Atendimento_Abs", "Cordialidade_Abs", "Dominio_Abs"]
)[["Funcionário", "Atendimento_Abs", "Cordialidade_Abs", "Dominio_Abs"]]

# Melt DataFrames for plotting
eficacia_melted_percent = eficacia_df_percent.melt(id_vars="Funcionário", var_name="Metric", value_name="Score")
eficacia_melted_abs = eficacia_df_abs.melt(id_vars="Funcionário", var_name="Metric", value_name="Score")

# Plot percentage graph
plt.figure(figsize=(14, 8))
sns.barplot(
    data=eficacia_melted_percent,
    x="Metric",
    y="Score",
    hue="Funcionário",
    palette=distinct_colors
)
plt.title("Avaliação dos Clientes em Percentual", fontsize=16, weight='bold')
plt.ylabel("Percentual (%)", fontsize=12)
plt.xlabel("Metricas", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Funcionário", fontsize=10)
plt.tight_layout()
plt.show()

# Plot absolute value graph
plt.figure(figsize=(14, 8))
sns.barplot(
    data=eficacia_melted_abs,
    x="Metric",
    y="Score",
    hue="Funcionário",
    palette=distinct_colors
)
plt.title("Avaliação dos Clientes valor absoluto", fontsize=16, weight='bold')
plt.ylabel("Contagem", fontsize=12)
plt.xlabel("Metricas", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Funcionário", fontsize=10)
plt.tight_layout()
plt.show()
```
